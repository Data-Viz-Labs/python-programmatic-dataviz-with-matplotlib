```python
# Dependencies

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
years = ["2004", "2022"]
norway  = [5, 8]
denmark = [4, 10]
sweden  = [13, 15]

x = np.arange(len(years))
width = 0.62
```

```python
# Colors
c_norway  = "#1f2a44"   # navy
c_denmark = "#f04e3c"   # red
c_sweden  = "#2f6fdd"   # blue
```

## We'll go with /#1



![viz-1.png](https://github.com/Data-Viz-Labs/python-programmatic-dataviz-with-matplotlib/blob/main/assets/viz-1.png?raw=true "sample #1")

```python
# With Matplotlib will be...

fig, ax = plt.subplots(figsize=(6, 6))

# Stacked bars
b1 = ax.bar(x, norway,  width, color=c_norway)
b2 = ax.bar(x, denmark, width, bottom=norway, color=c_denmark)
b3 = ax.bar(x, sweden,  width, bottom=np.array(norway)+np.array(denmark), color=c_sweden)

# Grid + axes styling
ax.set_axisbelow(True)
ax.yaxis.grid(True, alpha=0.25)
ax.xaxis.grid(False)
ax.set_xticks(x, years)
ax.set_ylim(0, 38)
for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)
ax.tick_params(axis="y", length=0)

# Labels inside each segment
def label_segments(bars, values, bottoms=None):
    if bottoms is None:
        bottoms = np.zeros(len(values))
    for rect, v, b in zip(bars, values, bottoms):
        ax.text(
            rect.get_x() + rect.get_width()/2,
            b + v/2,
            f"{v}",
            ha="center", va="center",
            color="white", fontsize=12
        )

label_segments(b1, norway)
label_segments(b2, denmark, bottoms=np.array(norway))
label_segments(b3, sweden,  bottoms=np.array(norway)+np.array(denmark))

# Totals on top
totals = np.array(norway) + np.array(denmark) + np.array(sweden)
for xi, t in zip(x, totals):
    ax.text(xi, t + 0.8, f"{t}", ha="center", va="bottom",
            fontsize=16, fontweight="bold")

# Right-side labels (as in the example)
x_text = x[-1] + 0.75
ax.text(x_text, 26, "Sweden",  color=c_sweden,  fontsize=18, va="center")
ax.text(x_text, 13.5, "Denmark", color=c_denmark, fontsize=18, va="center")
ax.text(x_text,  4.5, "Norway",  color=c_norway,  fontsize=18, va="center")

ax.margins(x=0.25)
plt.tight_layout()
plt.show()
```
